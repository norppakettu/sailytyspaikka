```python
bins = [250, 500, 750, 1000, 1250, 1500, 1750]
df['DailyRateClass'] = pd.cut(df['DailyRate'], bins = bins)

bins1 = [0, 5, 10, 15, 20, 25]
df['Distance Class'] = pd.cut(df['DistanceFromHome'], bins = bins1)

bins2 = [20, 40, 60, 80, 100, 120]
df['Hourly Rate Class'] = pd.cut(df['HourlyRate'], bins = bins2)

df.head()
```

```python
df.drop(['BusinessTravel', 'DailyRate', 'DistanceFromHome', 'Education',
         'HourlyRate', 'JobLevel', 'StockOptionLevel',
         'EmployeeCount', 'EmployeeNumber', 'WorkLifeBalance'], axis=1).head()
```

```python
df['Overall Satisfaction'] = df[['EnvironmentSatisfaction', 
                                 'JobSatisfaction']].mean(axis=1)

df[['Age', 'YearsAtCompany', 'Overall Satisfaction']].head()
```

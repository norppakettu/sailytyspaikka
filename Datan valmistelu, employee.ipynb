```python
df['Education_text'] = df['Education'].replace({1 : 'Below College', 
                                                2 : 'College',
                                                3 : 'Bachelor', 
                                                4 : 'Master', 
                                                5 : 'Doctor'})
df['JobInvolvement_text'] = df['JobInvolvement'].replace({ 1: 'Low',
                                                        2 : 'Medium',
                                                        3 : 'High',
                                                        4 : 'Very high'})
df['PerformanceRating_text'] = df['PerformanceRating'].replace({ 1: 'Low',
                                                              2 : 'Good',
                                                              3 : 'Excellent',
                                                              4 :'Outstanding'})
df['RelationshipSatisfaction_text'] = df['RelationshipSatisfaction'].replace({1 : 'Low', 
                                                                              2 : 'Medium',
                                                                              3 : 'High',
                                                                              4 :'Very high'})

df['WorkLifeBalance_text'] = df['WorkLifeBalance'].replace({1 : 'Bad',
                                                            2 : 'Good',
                                                            3 : 'Better', 
                                                            4 : 'Best'})
df.head()
```

```python
bins = [250, 500, 750, 1000, 1250, 1500, 1750]
df['DailyRateClass'] = pd.cut(df['DailyRate'], bins = bins)

bins1 = [0, 5, 10, 15, 20, 25]
df['Distance Class'] = pd.cut(df['DistanceFromHome'], bins = bins1)

bins3 = [10, 20, 30, 40, 50, 60]
df['Age Class'] = pd.cut(df['Age'], bins = bins3)

df.head()
```

```python
df['Overall Satisfaction'] = df[['EnvironmentSatisfaction', 
                                 'JobSatisfaction', 
                                 'RelationshipSatisfaction']].mean(axis=1)

df[['Age', 'YearsAtCompany', 'Overall Satisfaction']].head()
```

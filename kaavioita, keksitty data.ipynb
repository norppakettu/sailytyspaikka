```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.DataFrame({'Poikkeamat 2019': [55, 221, 245, 79, 115, 109]},
                 index= ['Tammikuu', 'Helmikuu', 'Maaliskuu',
                        'Huhtikuu', 'Toukokuu', 'Kesäkuu'])
df
```

```python
df.plot.bar()
```

```python
ax1 = df.plot.barh(legend=False, width=0.85, rot=10, color='orange')

ax1.set_title('Poikkeamat liikenteessä 2019')
ax1.set_xlabel('Lkm')
ax1.set_ylabel('Kuukausi')

ax1.grid(axis='y')
ax1.set_axisbelow(True)

plt.gcf().savefig('kuvio.png', bbox_inches='tight')
```

```python
df2 = pd.DataFrame({'Poikkeamat 2017': [22, 140, 260, 199, 20, 222],
                    'Poikkeamat 2018': [49, 201, 140, 90, 244, 50],
                    'Poikkeamat 2019': [55, 221, 245, 79, 115, 109]},
                 index= ['Tammikuu', 'Helmikuu', 'Maaliskuu',
                        'Huhtikuu', 'Toukokuu', 'Kesäkuu'])
df2
```

```python
ax2 = df2.plot.bar()

ax2.set_title('Poikkeamat liikenteessä 2017-2019')
ax2.set_xlabel('Kuukausi')
ax2.set_ylabel('Lkm')

ax2.tick_params(axis='x', labelrotation=0)
ax2.grid(axis='y')
ax2.set_axisbelow(True)

ax2.legend(loc=(-0.1, -0.25), ncol=3)
```

```python
ax2 = df2.T.plot.bar(width=0.7, figsize=(8,6))

ax2.set_xlabel('Vuosi')
ax2.set_ylabel('Lkm')

ax2.tick_params(axis='x', labelrotation=0, length=0)

ax2.grid(axis='y')

ax2.set_axisbelow(True)

ax2.legend(loc=(0.2, -0.25), ncol=3)
```

```python
ax3 = df2.plot(style='-o')

ax3.set_yticks([50,100,150,200,250])

ax3.set_ylabel('Lkm')

ax3.set_xticklabels(df2.index)

ax3.tick_params(axis='y', length=0)

ax3.grid()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

sns.set()
```

```python
wine = pd.read_excel('http://taanila.fi/wine.xlsx')
wine.head()
```

```python
X = wine.drop('Y', axis =1)
y= wine['Y']
```

```python
from sklearn.cluster import KMeans

malli = KMeans(n_clusters=3)
malli.fit(X)

malli.cluster_centers_
```

```python
X['K'] = malli.predict(X)
pd.crosstab(X['K'], 'total')
```

```python
X.groupby('K').describe().T
```

```python
pd.crosstab(X['K'], wine['Y'])
```

```python
import pandas as pd

df = pd.read_excel('valmisteltu512.xlsx')
df.head()
```

```python
df.drop(['Unnamed: 0'], axis=1).head()
```

```python
df1 = df.groupby('Education_text')['MonthlyIncome'].describe()

df1.style.format('{:.0f}')
```

```python
df2 = df.groupby(['Gender', 'Education_text'])['MonthlyIncome'].describe()

df2.style.format('{:.0f}')
```

```python
df3 = df.groupby(['Gender', 'Education_text'])['MonthlyIncome'].mean().unstack()

df3.style.format('{:.0f}')
```

```python
##tai pivotilla

dfp = df.pivot_table(values='MonthlyIncome', index='Gender', columns='Education_text')
dfp.style.format('{:.0f}')
```

```python
df4 = df.groupby(['Department', 'Education_text'])['MonthlyIncome', 'Overall Satisfaction'].mean()

df4.style.format({'MonthlyIncome':'{:.0f}', 'Overall Satisfaction':'{:.2f}'})
```

```python
df5 = df.groupby(['Department', 'Attrition'])['TrainingTimesLastYear', 'Overall Satisfaction'].mean()

df5.style.format({'TrainingTimesLastYear':'{:.2f}', 'Overall Satisfaction':'{:.2f}'})
```
